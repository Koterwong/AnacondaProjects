# 多元线性回归实战—房价预测

基于usa_housing_price.csv数据集，建立线性回归模型，预测房价

1. 先以面积size作为输入变量，建立一元线性回归模型，并评估模型表现，可视化其预测的结果
2. 以income、house age、numbers of rooms、popolation、size作为输入变量，建立多元线性回归模型，并评估模型表现
3. 预测：income=65000，house age=5，number of rooms=5，popolation=30000，size=200的房价

```python
# 引入相关的库
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 导入数据
data = pd.read_csv(r'usa_housing_price.csv')
data.head()
```

## 解答问题1：

```python
# 预展示数据集图形
fig = plt.figure(figsize=(22,22))

fig_1 = plt.subplot(231)
plt.scatter(data.loc[:,'Avg. Area Income'], data.loc[:,'Price'])
plt.title('Income VS Price')

fig_2 = plt.subplot(232)
plt.scatter(data.loc[:,'Avg. Area House Age'], data.loc[:,'Price'])
plt.title('House Age VS Price')

fig_3 = plt.subplot(233)
plt.scatter(data.loc[:,'Avg. Area Number of Rooms'], data.loc[:,'Price'])
plt.title('Number of Rooms Age VS Price')

fig_4 = plt.subplot(234)
plt.scatter(data.loc[:,'Area Population'], data.loc[:,'Price'])
plt.title('Area Population VS Price')

fig_5 = plt.subplot(235)
plt.scatter(data.loc[:,'size'], data.loc[:,'Price'])
plt.title('size Age VS Price')

plt.show()
```

```python
# 赋值
size = data.loc[:,'size']
price = data.loc[:,'Price']
```

```python
# 建立一元线性回归模型
from sklearn.linear_model import LinearRegression
```

```python
# 1维转换成2维，因为lr_model.fit接收参数是2维的
size = np.array(size).reshape(-1,1)
price = np.array(price).reshape(-1,1)
```

```python
lr_model = LinearRegression()
lr_model.fit(size,price) # 把数据喂给这个模型
```

```python
# 得出参数a，b
a = lr_model.coef_
b = lr_model.intercept_
print(a,b)
```

```python
# 模型评估
from sklearn.metrics import mean_squared_error,r2_score
price_predict = lr_model.predict(size) # 预测结果
MSE = mean_squared_error(price,price_predict) # 均方差，越接近0越好
R2 = r2_score(price,price_predict) # R方值，越接近1越好
print(MSE,R2) # 预测结果不理想
```

```python
fig_6 = plt.figure(figsize=(8,8))
plt.scatter(price,price_predict)
plt.show() # 点还是太散了，拟合效果不理想
```

结论：从结果可以看出，拟合程度不算很好，MSE和R2的值都不好，所以接下来要建立多元线性回归模型

## 问题2：

```python
# 赋值
X = data.drop(['Price'],axis=1) # 取除了Price的其他列
print(np.array(X))
X
```

```python
# 模型训练
LR_model = LinearRegression()
LR_model.fit(X,price) # 把数据喂给这个模型
```

```python
# 模型表现评估
Y_predict = LR_model.predict(X)
MSE = mean_squared_error(price,Y_predict) # 均方差，越接近0越好
R2 = r2_score(price,Y_predict) # R方值，越接近1越好
print(MSE,R2) # 效果相对一元线性模型好了很多
```

```python
fig_7 = plt.figure(figsize=(8,8))
plt.scatter(price,Y_predict)
plt.show() # 点没那么散了
```

结果：多元线性模型看起来拟合程度比一元线性模型的好了很多
