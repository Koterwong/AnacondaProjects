# 考试通过预测实战
1. 基于examdata.csv数据集，建立逻辑回归模型，并评估模型表现
2. 预测Exam1=75，Exam2=60时，该同学能否通过Exam3
3. 为了有更好的模型表现，建立二阶边界函数，重复任务1、2

## 任务1
一阶边界函数： $\theta_0 + \theta_1 X_1 + \theta_2 X_2 = 0$
![linear_boundary.png](images/linear_boundary.png)

```python
# 导入相关库
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 导入数据集
data = pd.read_csv(r'examdata.csv')
data.head()
```

```python
# 数据预可视化
plt.figure(figsize=(8,6))
plt.scatter(data.loc[:,'Exam1'], data.loc[:,'Exam2'])
plt.title('Exam1-Exam2')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.show()
```

```python
# 添加标记加以区分
mask = data.loc[:,'Pass'] == 1
plt.figure(figsize=(8,6))
passed = plt.scatter(data.loc[:,'Exam1'][mask], data.loc[:,'Exam2'][mask]) # 通过
failed = plt.scatter(data.loc[:,'Exam1'][~mask], data.loc[:,'Exam2'][~mask]) # 不通过
plt.title('Exam1-Exam2')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.legend((passed,failed),('passed','failed'))
plt.show()
```

```python
# 定义数据
x = data.drop(['Pass'],axis=1)
y = data.loc[:,'Pass']
x1 = data.loc[:,'Exam1']
x2 = data.loc[:,'Exam2']
```

```python
# 模型训练
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x,y)
```

```python
# 评估模型表现
from sklearn.metrics import accuracy_score
y_predict = LR.predict(x)
accurancy = accuracy_score(y,y_predict)
print(accurancy) # 准确率为0.89，算比较高的
```

```python
# 可视化一阶边界曲线

# 先得出3个参数theta0、theta1、theta2
theta0 = LR.intercept_
theta1, theta2 = LR.coef_[0][0], LR.coef_[0][1]

# 根据一阶边界函数，得
x2_new = -(theta0 + theta1 * x1) / theta2

# 绘图
plt.figure(figsize=(8,6))
passed = plt.scatter(data.loc[:,'Exam1'][mask], data.loc[:,'Exam2'][mask])
failed = plt.scatter(data.loc[:,'Exam1'][~mask], data.loc[:,'Exam2'][~mask])
plt.plot(x1,x2_new)
plt.title('Exam1-Exam2')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.legend((passed,failed),('passed','failed'))
plt.show()
```

结论：这条一阶边界曲线准确率还可以，不过还有一些点没有正确分类，接下来就使用二阶边界函数

## 任务2

```python
x_test = [[75,60]]
y_test_predict = LR.predict(x_test)
print('passed' if y_test_predict == 1 else 'failed') # 结果为1，通过
```

## 任务3

二阶边界函数：$\theta_0 + \theta_1 X_1 + \theta_2 X_2+ \theta_3 X_1^2 + \theta_4 X_2^2 + \theta_5 X_1 X_2 = 0$

转换一下：

- $a x^2 + b x + c=0: x1 = (-b+(b^2-4ac)^.5)/2a,x1 = (-b-(b^2-4ac)^.5)/2a$

- $\theta_4 X_2^2 + (\theta_5 X_1++ \theta_2) X_2 + (\theta_0 + \theta_1 X_1 + \theta_3 X_1^2)=0$

```python
x1_2 = x1 * x1
x2_2 = x2 * x2
x1_x2 = x1 * x2
x_new = {'x1': x1,'x2': x2,'x1_2': x1_2,'x2_2': x2_2,'x1_x2': x1_x2}
x_new = pd.DataFrame(x_new)
print(x_new)
```

```python
# 模型训练
LR2 = LogisticRegression()
LR2.fit(x_new,y)
```

```python
# 评估模型表现
y2_predict = LR2.predict(x_new)
accuracy2 = accuracy_score(y,y2_predict)
print(accuracy2) # 准确率为1，完美
```

```python
# 可视化二阶边界曲线

# 先排序
x1_new = x1.sort_values()

# 得出参数theta0、theta1、theta2、theta3、theta4、theta5
theta0 = LR2.intercept_
theta1,theta2,theta3,theta4,theta5 = LR2.coef_[0][0],LR2.coef_[0][1],LR2.coef_[0][2],LR2.coef_[0][3],LR2.coef_[0][4]

# 根据二阶边界函数，得
a = theta4
b = theta5*x1_new+theta2
c = theta0+theta1*x1_new+theta3*x1_new*x1_new
x2_new_boundary = (-b+np.sqrt(b*b-4*a*c))/(2*a)

# 绘图
plt.figure(figsize=(12,10))
passed=plt.scatter(data.loc[:,'Exam1'][mask],data.loc[:,'Exam2'][mask])
failed=plt.scatter(data.loc[:,'Exam1'][~mask],data.loc[:,'Exam2'][~mask])
plt.plot(x1_new,x2_new_boundary)
plt.title('Exam1-Exam2')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.legend((passed,failed),('passed','failed'))
plt.show()
```

结论：二阶边界函数比一阶边界函数，效果更好
